# COVID-19

```python
import pandas as pd
import seaborn as sns; sns.set()
```

```python
death = pd.read_csv('./data/deaths_cumulative_daily.csv')
death['Date'] = pd.to_datetime(death['Date'], format='%Y/%m/%d')
death
```

## Visualize Cumulative death toll

```python
sns.lineplot(data=death, x='Date', y='ALL')
```

## Visualize Number of deaths on the day

```python
sns.lineplot(
    data=pd.concat([death['Date'], death['ALL'].diff()], axis=1),
    x='Date',
    y='ALL'
)
```

```python
positive = pd.read_csv('./data/confirmed_cases_cumulative_daily.csv')
positive['Date'] = pd.to_datetime(positive['Date'], format='%Y/%m/%d')
positive
```

## Visualize Cumulative number of infected people

```python
sns.lineplot(data=positive, x='Date', y='ALL')
```

```python
sns.lineplot(
    data=pd.concat([positive['Date'], positive['ALL'].diff()], axis=1),
    x='Date',
    y='ALL'
)
```
